```python
df.set_index('Name')
```

```python
i_v = df.loc[df['Name'] == 'Iris-virginica']
i_s = df.loc[df['Name'] == 'Iris-setosa']
i_c = df.loc[df['Name'] == 'Iris-versicolor']
```

```python
#6.4 Create a histogram for each feature in the dataset to illustrate the feature distributions, using the default number of bins. A histogram provides a graphical representation of the distribution of the data.

%matplotlib inline
import numpy as np 



plt.hist([i_s['Sepal Length'],i_v['Sepal Length'],i_c['Sepal Length']], label=('Setosa','Virginica','Versicolor'))
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.legend()
plt.show()

plt.hist([i_s['Sepal Width'],i_v['Sepal Width'],i_c['Sepal Width']], label=('Setosa','Virginica','Versicolor'))
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.legend()
plt.show()

plt.hist([i_s['Petal Length'],i_v['Petal Length'],i_c['Petal Length']], label=('Setosa','Virginica','Versicolor'))
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.legend()
plt.show()

plt.hist([i_s['Petal Width'],i_v['Petal Width'],i_c['Petal Width']], label=('Setosa','Virginica','Versicolor'))
plt.xlabel('Petal Width')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
#6.5 Create a boxplot for each feature in the dataset. All of the boxplots will be combined into a single plot. A boxplot provides a graphical representation of the location and variation of the data through their quartiles; they are especially useful for comparing distributions and identifying outliers.


plt.boxplot([i_s['Sepal Length'],i_v['Sepal Length'],i_c['Sepal Length']])
plt.xlabel('Sepal Length')
plt.xticks([1,2,3],['Setosa','Virginica','Versicolor'])
plt.ylabel('Frequency')
plt.show()

plt.boxplot([i_s['Sepal Width'],i_v['Sepal Width'],i_c['Sepal Width']])
plt.xlabel('Sepal Width')
plt.xticks([1,2,3],['Setosa','Virginica','Versicolor'])
plt.ylabel('Frequency')
plt.show()

plt.boxplot([i_s['Petal Length'],i_v['Petal Length'],i_c['Petal Length']])
plt.xlabel('Petal Length')
plt.xticks([1,2,3],['Setosa','Virginica','Versicolor'])
plt.ylabel('Frequency')
plt.show()

plt.boxplot([i_s['Petal Width'],i_v['Petal Width'],i_c['Petal Width']])
plt.xlabel('Petal Width')
plt.xticks([1,2,3],['Setosa','Virginica','Versicolor'])
plt.ylabel('Frequency')
plt.show()
```
